```python
from cmtj.llgb import MFAWeissCurie
import matplotlib.pyplot as plt
from collections import defaultdict
import numpy as np
import pandas as pd

J0 = 3.051823739e-20
me = 0.2

scans = {}
Tsrange = np.linspace(1, 1000, 200)
for J0 in (1.e-20, 2.e-20, 3e-20):
    data = []
    for Ts in Tsrange:
        res = MFAWeissCurie(me, Ts, J0)
        est = res[0]
        data.append(est)
    scans[J0] = data
```

```python
colors = ['crimson', 'royalblue', 'forestgreen']
with plt.style.context(["science", "nature"]):
    fig, ax = plt.subplots(dpi=300)
    for i, (k, v) in enumerate(scans.items()):
        ax.plot(Tsrange,
                v,
                label=f"$J_0=$" + str(int(k * 1e20)),
                color=colors[i])
    ax.set_xlabel("Temperature (K)")
    ax.set_ylabel("$m_e$")
    ax.legend()
```
